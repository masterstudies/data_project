```python
from AsyncParse.request import RequestManager
import asyncio
import pandas as pd
```

```python
df = pd.read_excel("full__dataset_progs.xlsx")
links = df['Link'].tolist()
```

```python
links 
```

```python
loop = asyncio.get_event_loop()
request_manager = RequestManager(
    loop = loop,
    limit=100,
    rate=1,
    timeout=5
)
```

```python
data = request_manager.scrape_urls(links)
data
```
